```python
import numpy as np

X_train = np.load('/content/drive/MyDrive/1. Natural Language Processsing/Khóa luận tốt nghiệp/Word2vec/English/training.npy')
y_train = np.load('/content/drive/MyDrive/1. Natural Language Processsing/Khóa luận tốt nghiệp/Word2vec/English/training_labels.npy')
X_test = np.load('/content/drive/MyDrive/1. Natural Language Processsing/Khóa luận tốt nghiệp/Word2vec/English/test.npy')
y_test = np.load('/content/drive/MyDrive/1. Natural Language Processsing/Khóa luận tốt nghiệp/Word2vec/English/test_labels.npy')
```

```python
from sklearn.ensemble import RandomForestClassifier

text_classifier = RandomForestClassifier(n_estimators=100, random_state=0)
text_classifier.fit(X_train, y_train)
```

```python
predictions_rf = text_classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test,predictions_rf))
print(classification_report(y_test,predictions_rf))
print(accuracy_score(y_test, predictions_rf))
```

```python
from sklearn.tree import DecisionTreeClassifier

text_classifier = DecisionTreeClassifier()
text_classifier.fit(X_train, y_train)
```

```python
predictions_dt = text_classifier.predict(X_test)
print(confusion_matrix(y_test,predictions_dt))
print(classification_report(y_test,predictions_dt))
print(accuracy_score(y_test, predictions_dt))
```

```python
from sklearn.linear_model import LogisticRegression

text_classifier = LogisticRegression()
text_classifier.fit(X_train, y_train)

predictions_lr = text_classifier.predict(X_test)
print(confusion_matrix(y_test,predictions_lr))
print(classification_report(y_test,predictions_lr))
print(accuracy_score(y_test, predictions_lr))
```

```python
from sklearn.ensemble import AdaBoostClassifier

text_classifier = AdaBoostClassifier(random_state=42)
text_classifier.fit(X_train, y_train)

predictions_ada = text_classifier.predict(X_test)
print(confusion_matrix(y_test,predictions_ada))
print(classification_report(y_test,predictions_ada))
print(accuracy_score(y_test, predictions_ada))
```

```python
from sklearn.naive_bayes import GaussianNB

text_classifier = GaussianNB()
text_classifier.fit(X_train, y_train)

predictions_NB = text_classifier.predict(X_test)
print(confusion_matrix(y_test,predictions_NB))
print(classification_report(y_test,predictions_NB))
print(accuracy_score(y_test, predictions_NB))
```

```python
from sklearn.neighbors import KNeighborsClassifier

text_classifier = KNeighborsClassifier(7)
text_classifier.fit(X_train, y_train)

predictions_knn = text_classifier.predict(X_test)
print(confusion_matrix(y_test,predictions_knn))
print(classification_report(y_test,predictions_knn))
print(accuracy_score(y_test, predictions_knn))
```

```python
from sklearn.neural_network import MLPClassifier

text_classifier = MLPClassifier()
text_classifier.fit(X_train, y_train)

predictions_mlp = text_classifier.predict(X_test)
print(confusion_matrix(y_test,predictions_mlp))
print(classification_report(y_test,predictions_mlp))
print(accuracy_score(y_test, predictions_mlp))
```

```python
from sklearn.svm import SVC

text_classifier = SVC()
text_classifier.fit(X_train, y_train)

predictions_svc = text_classifier.predict(X_test)
print(confusion_matrix(y_test,predictions_svc))
print(classification_report(y_test,predictions_svc))
print(accuracy_score(y_test, predictions_svc))
```

```python
from xgboost import XGBClassifier

text_classifier = XGBClassifier()
text_classifier.fit(X_train, y_train)

predictions_xgb = text_classifier.predict(X_test)
print(confusion_matrix(y_test,predictions_xgb))
print(classification_report(y_test,predictions_xgb))
print(accuracy_score(y_test, predictions_xgb))
```

```python
import matplotlib.pyplot as plt

models = ['RF', 'DT', 'LR', 'Ada', 'NB', 'KNN', 'MLP', 'SVC', 'XGB']
acc = [accuracy_score(y_test, predictions_rf),
       accuracy_score(y_test, predictions_dt),
       accuracy_score(y_test, predictions_lr),
       accuracy_score(y_test, predictions_ada),
       accuracy_score(y_test, predictions_NB),
       accuracy_score(y_test, predictions_knn),
       accuracy_score(y_test, predictions_mlp),
       accuracy_score(y_test, predictions_svc),
       accuracy_score(y_test, predictions_xgb)]


plt.bar(models, acc)
plt.grid()
plt.show()
```
